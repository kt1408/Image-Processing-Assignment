```python
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
import cv2
from sklearn import datasets, metrics
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier as KNN

!wget https://i.imgur.com/Y0sNjp5.png
!mv Y0sNjp5.png lena.png
```

```python
# 1. Histogram equaliaztion
def equalize(image):
    img = np.asarray(image.convert('L'))
    width, height = img.shape
    h = np.bincount(img.flatten())
    h_ = np.copy(h)
    for i in range(1, 256):
        h_[i] += h_[i - 1] # prefix sums

    g = np.floor(h_ * 255 / h_[-1]).astype(np.uint8) # map of pixels

    eq_img = np.zeros((width, height), dtype=np.uint8)
    for i in range(width):
        for j in range(height):
            eq_img[i, j] = g[img[i, j]]
    
    fig, ax = plt.subplots(2, 2, figsize=(16, 16))
    ax[0, 0].set_title('Original image')
    ax[0, 0].imshow(img, cmap='gray')
    ax[1, 0].set_title('Transformed image')
    ax[1, 0].imshow(eq_img, cmap='gray')
    ax[0, 1].plot(range(0, 256), h)
    ax[1, 1].plot(range(0, 256), np.bincount(eq_img.flatten()))

    return eq_img

equalized_img = equalize(Image.open('./lena.png'))
```

```python
# 2. DFT + LPF/HPF
def hpf(image, cutoff):
    img = np.asarray(image.convert('L'))
    width, height = img.shape

    dft = np.fft.fft2(img) # forward transform
    shifted = np.fft.fftshift(dft)

    lpfilter = np.zeros((width, height)) # low pass filter
    cv2.circle(lpfilter, (width // 2, height // 2), cutoff, (255, 255, 255), -1)
    hpfilter = 255 - lpfilter # high pass filter => (1 - low pass)

    low_pass = np.multiply(lpfilter, shifted) / 255
    low_pass = np.fft.ifftshift(low_pass)
    low_pass = np.fft.ifft2(low_pass) # inverse transform
    low_pass = np.abs(low_pass).clip(0, 255).astype(np.uint8)
    
    high_pass = np.multiply(hpfilter, shifted) / 255
    high_pass = np.fft.ifftshift(high_pass)
    high_pass = np.fft.ifft2(high_pass) # inverse transform
    high_pass = np.abs(high_pass).clip(0, 255).astype(np.uint8)

    fig, (lp, hp) = plt.subplots(1, 2, figsize=(10, 5))
    lp.imshow(low_pass, cmap='gray')
    lp.set_title('Low pass image')
    hp.imshow(high_pass, cmap='gray')
    hp.set_title('High pass image')

for d0 in [2, 4, 8, 16, 32, 64, 128, 512]: # for different cutoff freqs
    hpf(Image.open('./lena.png'), d0)
```

```python
#3. Image segmentation
def image_segment(img, k=3):
    img = cv2.imread(img)
    width, height, channels = img.shape

    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = img.reshape((-1,3)).astype(np.float32)

    error, labels, centers = cv2.kmeans(img, k, None,
                (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 5, 0.95),
                3, cv2.KMEANS_RANDOM_CENTERS)

    segmented_img = np.zeros((width, height, channels))

    colors = [tuple(np.random.randint(256) for _ in range(channels))
                for _ in range(k)]

    for i in range(0, width):
        for j in range(0, height):
            segmented_img[i,j] = colors[labels[i * width + j][0]]
    segmented_img = segmented_img.astype(np.uint8)
    plt.imshow(segmented_img)
    plt.figure()
    return error


plt.imshow(np.asarray(Image.open('./lena.png')))
plt.title('Original image')
plt.show()
plt.figure()

max_k = 6
k_range = range(1, max_k + 1)
error = np.zeros(max_k)
for k in k_range:
    error[k - 1] = image_segment('./lena.png', k)

plt.title('Error vs K')
plt.plot(k_range, error)
plt.show()
```

```python
#4. Using MNIST for KNN Classification
def knn_classification_mnist(k=3):
    digits = datasets.load_digits()
    data = digits.images.reshape((len(digits.images), -1))
    clf = KNN(k)

    x_train, x_test, y_train, y_test = train_test_split(data, 
                                                digits.target, test_size=0.3)
    clf.fit(x_train, y_train)
    predicted = clf.predict(x_test)

    print(f'Train accuracy: \
{metrics.accuracy_score(y_train, clf.predict(x_train)) * 100:.2f}%', )
    print(f'Test accuracy: \
{metrics.accuracy_score(y_test, predicted) * 100:.2f}%')
    print(metrics.classification_report(y_test, predicted))
    disp = metrics.ConfusionMatrixDisplay.from_predictions(y_test, predicted)
    plt.show()

knn_classification_mnist()
```
